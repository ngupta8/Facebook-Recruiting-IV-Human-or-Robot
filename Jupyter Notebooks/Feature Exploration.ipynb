### Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from rfpimp import *
```

__Uploading the file withh features__

```python
feature_of_all_train_df = pd.read_csv('../DataSet/train_data_with_features.csv')
```

```python
feature_of_all_train_df.head()
```

__Heat Map of correlation of all features__

```python
plt.figure(figsize=(10,8))
sns.heatmap(feature_of_all_train_df.corr(),cmap='Blues',annot=False)
plt.savefig('heatmap') 
```

__Correlation of Quality correlation matrix__

```python
k = 23 #number of variables for heatmap
cols = feature_of_all_train_df.corr()['outcome'].index
cm = feature_of_all_train_df[cols].corr()
plt.figure(figsize=(15,12))
sns.heatmap(cm, annot=True, cmap = 'viridis') 
plt.savefig('matrix') 
```

 __Droping the features which are highly correlated and not needed__

```python
feature_of_all_train_df_clean = feature_of_all_train_df.drop('merc_count', axis=1)
feature_of_all_train_df_clean.drop('average_distinct_ip_per_auction', axis=1, inplace=True)
feature_of_all_train_df_clean.drop('average_distinct_urls_per_auction', axis=1, inplace=True)
feature_of_all_train_df_clean.drop('auction_count', axis=1, inplace=True)
feature_of_all_train_df_clean.drop('device_count', axis=1, inplace=True)
feature_of_all_train_df_clean.drop('median_response_time', axis=1, inplace=True)
feature_of_all_train_df_clean.drop('median_response_time_across_auction', axis=1, inplace=True)
feature_of_all_train_df_clean.drop('median_time_to_respond_competition', axis=1, inplace=True)
feature_of_all_train_df_clean.drop('median_time_to_respond_self', axis=1, inplace=True)
feature_of_all_train_df_clean.drop('url_count', axis=1, inplace=True)

k = 22 #number of variables for heatmap
cols = feature_of_all_train_df_clean.corr().nlargest(k, 'outcome')['outcome'].index
cm = feature_of_all_train_df[cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, cmap = 'viridis')
plt.savefig('after feature pruning.png') 
```

__Checking distribution of features__

```python
l = feature_of_all_train_df_clean.iloc[:,1:]
l = l.columns.values
number_of_columns= 7
number_of_rows = (len(l)+1)//number_of_columns
fig = plt.figure(figsize=(15,8))
fig.subplots_adjust(hspace=1, wspace=0.6)
for i in range(1,len(l)):
    plt.subplot(number_of_rows,number_of_columns,i)
    sns.set_style('whitegrid')
    b = sns.boxplot(feature_of_all_train_df_clean[l[i-1]],color='green',orient='v')
    b.set_ylabel(l[i-1],fontsize=15)
    b.tick_params(labelsize=10)
    plt.tight_layout()
plt.savefig('merc_count_auction_distribution.png') 
```
